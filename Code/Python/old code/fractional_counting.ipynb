```python
#select just project_id and progr_proj_id
progr_proj_id = data[['project_id', 'progr_proj_id']]
progr_proj_id = progr_proj_id.sort_values(by=['project_id'], ascending=True)
progr_proj_id = progr_proj_id.drop_duplicates()

#clean progr_proj_id from H2020 and remove non-alphanumeric chars 
progr_proj_id['progr_proj_id'] = progr_proj_id['progr_proj_id'].str.replace('H2020', '')
progr_proj_id['progr_proj_id'] = progr_proj_id['progr_proj_id'].str.replace('\W', ' ')

#strip the string on the left and on the right (remove spaces in the beginning and the end of the string)
progr_proj_id['progr_proj_id'] = progr_proj_id['progr_proj_id'].str.lstrip() 
progr_proj_id['progr_proj_id'] = progr_proj_id['progr_proj_id'].str.rstrip()
```

```python
progr_proj_id
```

```python
i = progr_proj_id.columns.get_loc('progr_proj_id')
xx = progr_proj_id['progr_proj_id'].str.split(" ", expand=True)
xx2 = pd.concat([progr_proj_id.iloc[:, :i], xx, progr_proj_id.iloc[:, i+1:]], axis=1)
xx2 = xx2.rename(columns={0: 'progr_proj_id_1', 1: 'progr_proj_id_2'})
xx2 = xx2[['project_id', 'progr_proj_id_1', 'progr_proj_id_2']]
xx2
```

```python
#restructure the dataframe so that each project has a row for each project_id
progr_proj_id_1 = xx2[['project_id', 'progr_proj_id_1']]
progr_proj_id_1 = progr_proj_id_1.rename(columns={'progr_proj_id_1': 'progr_proj_id'})
progr_proj_id_2 = xx2[['project_id', 'progr_proj_id_2']]
progr_proj_id_2 = progr_proj_id_2.rename(columns={'progr_proj_id_2': 'progr_proj_id'})

xx3 = progr_proj_id_1.append(progr_proj_id_2)
xx3 = xx3.drop_duplicates()
xx3 = xx3.sort_values(by=['project_id'], ascending=True)
xx3 = xx3.dropna()

#clean progr_proj_id
xx3 = xx3.loc[(xx3.progr_proj_id != 'Manufacturing') & 
                (xx3.progr_proj_id != 'Space') &
                (xx3.progr_proj_id != 'Materials') &
                (xx3.progr_proj_id != 'Biotech') &
                (xx3.progr_proj_id != 'in') &
                (xx3.progr_proj_id != 'Nano') &
                (xx3.progr_proj_id != 'Innvoation') &
                (xx3.progr_proj_id !=  '0')]

xx3
```

```python
#get dummies for progr_proj_id
dummies_progr_proj_id = pd.get_dummies(xx3, columns=['progr_proj_id'])
dummies_progr_proj_id = dummies_progr_proj_id.groupby(['project_id']).max().reset_index()
dummies_progr_proj_id = dummies_progr_proj_id.sort_values(by=['project_id'], ascending=True)
dummies_progr_proj_id
```

```python
#select just project_id and progr_proj_id
action_type_proj_id = data[['project_id', 'action_type_proj_id']]
action_type_proj_id = action_type_proj_id.sort_values(by=['project_id'], ascending=True)
action_type_proj_id = action_type_proj_id.drop_duplicates()
action_type_proj_id
```

```python
i = action_type_proj_id.columns.get_loc('action_type_proj_id')
xx = action_type_proj_id['action_type_proj_id'].str.split("-", expand=True)
xx2 = pd.concat([action_type_proj_id.iloc[:, :i], xx], axis=1)
xx2 = xx2.rename(columns={0: 'action_type_proj_id_1', 1: 'action_type_proj_id_2', 2: 'action_type_proj_id_3', 3: 'action_type_proj_id_4'})
#xx2 = xx2[['project_id', 'progr_proj_id_1', 'progr_proj_id_2']]
xx2
```

```python
#restructure the dataframe so that each project has a row for each project_id
action_type_proj_id_1 = xx2[['project_id', 'action_type_proj_id_1']]
action_type_proj_id_1 = action_type_proj_id_1.rename(columns={'action_type_proj_id_1': 'action_type_proj_id'})

action_type_proj_id_2 = xx2[['project_id', 'action_type_proj_id_2']]
action_type_proj_id_2 = action_type_proj_id_2.rename(columns={'action_type_proj_id_2': 'action_type_proj_id'})

action_type_proj_iD_3 = xx2[['project_id', 'action_type_proj_id_3']]
action_type_proj_iD_3 = action_type_proj_iD_3.rename(columns={'action_type_proj_id_3': 'action_type_proj_id'})

action_type_proj_id_4 = xx2[['project_id', 'action_type_proj_id_4']]
action_type_proj_id_4 = action_type_proj_id_4.rename(columns={'action_type_proj_id_4': 'action_type_proj_id'})

xx3 = action_type_proj_id_1.append(action_type_proj_id_2)
xx3 = xx3.append(action_type_proj_iD_3)
xx3 = xx3.append(action_type_proj_id_4)

xx3 = xx3.drop_duplicates()
xx3 = xx3.sort_values(by=['project_id'], ascending=True)
xx3 = xx3.dropna()

#clean action_type_proj_id
xx3 = xx3.loc[(xx3.action_type_proj_id != '2') & 
                (xx3.action_type_proj_id != '2b')]

xx3
```

```python
#get dummies for action_type_proj_id
dummies_action_type_proj_id = pd.get_dummies(xx3, columns=['action_type_proj_id'])
dummies_action_type_proj_id = dummies_action_type_proj_id.groupby(['project_id']).max().reset_index()
dummies_action_type_proj_id = dummies_action_type_proj_id.sort_values(by=['project_id'], ascending=True)
dummies_action_type_proj_id
```
