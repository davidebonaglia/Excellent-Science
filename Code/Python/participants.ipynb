```python
import pandas as pd

data_dir = '/Users/davidebonaglia/Dropbox/Projects UNU/Excellent Science/'
participants = data_dir + 'Participants_data_combined.xlsx'

participants = pd.read_excel(participants)
participants
```

```python
#select just a few columns
participants = participants[['CD_PROJ_ID', 'CD_APPL_PIC', 'CD_PART_NR', 'AM_COST', 
                            'AM_EC_CONTRIB', 'AM_NET_CONTRIB', 'CD_ORG_NUTS_CODE', 'CD_LATITUDE', 'CD_LONGITUDE',
                            'CD_PROJ_ORG_TYPE*', 'cd_part_nuts', 'fl_sme', 'cd_org_type*', 'orbis_size_classification',
                            'technote_industry', 'applications_number_funded', 'applications_total_number', 
                            'applications_number_above_treshold', 'applications_success_rate_']]

#drop NaN 
participants = participants.dropna()

#declare some variables as integers
participants['CD_PROJ_ID'] = participants['CD_PROJ_ID'].astype(int)
participants['CD_APPL_PIC'] = participants['CD_APPL_PIC'].astype(int)
participants['CD_PART_NR'] = participants['CD_PART_NR'].astype(int)
participants['AM_COST'] = participants['AM_COST'].astype(int)
participants['AM_EC_CONTRIB'] = participants['AM_EC_CONTRIB'].astype(int)
participants['AM_NET_CONTRIB'] = participants['AM_NET_CONTRIB'].astype(int)
participants['applications_number_funded'] = participants['applications_number_funded'].astype(int)
participants['applications_total_number'] = participants['applications_total_number'].astype(int)
participants['applications_number_above_treshold'] = participants['applications_number_above_treshold'].astype(int)
participants['applications_success_rate_'] = participants['applications_success_rate_'].astype(int)

#rename column of project ID
participants = participants.rename(columns={'CD_PROJ_ID': 'project_id', 'CD_PROJ_ORG_TYPE*':'cd_org_type'})
participants
```

```python
########################################################################################
#cd_org_type
```

```python
cd_org_type = participants[['project_id', 'CD_APPL_PIC', 'cd_org_type']]

cd_org_type = cd_org_type.sort_values(by=['project_id'], ascending=True)
cd_org_type
```

```python
#get dummies for cd_org_type
dummies_cd_org_type = pd.get_dummies(cd_org_type, columns=['cd_org_type'])

dummies_cd_org_type = dummies_cd_org_type.loc[:, ~dummies_cd_org_type.columns.isin(['CD_APPL_PIC'])]
dummies_cd_org_type = dummies_cd_org_type.groupby(['project_id']).max().reset_index()
dummies_cd_org_type = dummies_cd_org_type.sort_values(by=['project_id'], ascending=True)
dummies_cd_org_type
```

```python
###########################################################################
#project characteristics

cd_and_costs = participants[['project_id', 'CD_APPL_PIC', 'CD_PART_NR', 'AM_COST', 'AM_EC_CONTRIB', 'AM_NET_CONTRIB']]
cd_and_costs
```

```python
import numpy as np

cd_and_costs['total_cost'] = np.where(cd_and_costs.AM_COST !=0, cd_and_costs.AM_COST, cd_and_costs.AM_EC_CONTRIB)
cd_and_costs
```

```python
proj_character = cd_and_costs.groupby(['project_id']).agg({'CD_APPL_PIC':'nunique',
                                                           'CD_PART_NR': 'sum',
                                                           'total_cost': 'sum'}).reset_index()

proj_character = proj_character.rename(columns={'CD_APPL_PIC': 'n_organizations',
                                                'CD_PART_NR': 'n_participants',
                                                'total_cost': 'total_cost_proj'})

proj_character
```

```python
###########################################################################
#country of the participant
```

```python
#consider just a few columns and sort values by CD_APPL_PIC 
countries = participants[['project_id', 'CD_APPL_PIC', 'CD_ORG_NUTS_CODE']]
countries = countries.sort_values(by=['project_id'], ascending=True) 


#clean CD_ORG_NUTS_CODE column by selecting the firsty 2 characters
countries['CD_ORG_NUTS_CODE'] = countries['CD_ORG_NUTS_CODE'].str[:2]
countries = countries.drop_duplicates()
countries = countries.rename(columns={'CD_ORG_NUTS_CODE': 'country'})

countries
```

```python
#EU27 = Austria, Belgium, Bulgaria, Croatia, 
#Republic of Cyprus, Czech Republic, Denmark, 
#Estonia, Finland, France, Germany, Greece, Hungary, 
#Ireland, Italy, Latvia, Lithuania, Luxembourg, Malta, 
#Netherlands, Poland, Portugal, Romania, Slovakia, Slovenia, Spain and Sweden.

#A: Widening EU countries:
#Bulgaria, Croatia, Cyprus, Czech Republic, Estonia, 
#Hungary, Latvia, Lithuania, Luxembourg, Malta, Poland, Portugal, Romania, Slovakia, Slovenia
```

```python
import numpy as np

countries['US'] = np.where((countries.country == 'US'), 1, 0)
countries['UK'] = np.where((countries.country == 'UK'), 1, 0)
countries['KR'] = np.where((countries.country == 'KR'), 1, 0)
countries['JP'] = np.where((countries.country == 'JP'), 1, 0)
countries['CH'] = np.where((countries.country == 'CH'), 1, 0)

countries['EU27'] = np.where((countries.country == 'AT') |
                            (countries.country == 'BE') |
                            (countries.country == 'BG') |
                            (countries.country == 'HR') |
                            (countries.country == 'CY') |
                            (countries.country == 'CZ') |
                            (countries.country == 'DK') |
                            (countries.country == 'EE') |
                            (countries.country == 'FI') |
                            (countries.country == 'FR') |
                            (countries.country == 'DE') |
                            (countries.country == 'EL') |
                            (countries.country == 'HU') |
                            (countries.country == 'IE') |
                            (countries.country == 'IT') |
                            (countries.country == 'LV') |
                            (countries.country == 'LT') |
                            (countries.country == 'LU') |
                            (countries.country == 'MT') |
                            (countries.country == 'NL') |
                            (countries.country == 'PL') |
                            (countries.country == 'PT') |
                            (countries.country == 'RO') |
                            (countries.country == 'SK') |
                            (countries.country == 'SI') |
                            (countries.country == 'ES') |
                            (countries.country == 'SE'), 1, 0)


#A: Widening EU countries:
#Bulgaria, Croatia, Cyprus, Czech Republic, Estonia, 
#Hungary, Latvia, Lithuania, Luxembourg, Malta, Poland, Portugal, Romania, Slovakia, Slovenia

countries['EU_Widening'] = np.where((countries.country == 'BG') |
                                    (countries.country == 'HR') |
                                    (countries.country == 'CY') |
                                    (countries.country == 'CZ') |
                                    (countries.country == 'EE') |
                                    (countries.country == 'HU') |
                                    (countries.country == 'LV') |
                                    (countries.country == 'LT') |
                                    (countries.country == 'LU') |
                                    (countries.country == 'MT') |
                                    (countries.country == 'PL') |
                                    (countries.country == 'PT') |
                                    (countries.country == 'RO') |
                                    (countries.country == 'SI') |
                                    (countries.country == 'SK'), 1, 0)

#B: Associated countries (non-EU):
#Albania, Armenia, Bosnia and Herzegovina, Faroe Islands, 
# #Georgia, Iceland, Israel, Kosovo, Moldova, Montenegro, 
# #North Macedonia, Norway, Serbia, Tunisia, Turkey, Ukraine

countries['not_EU_Widening'] = np.where((countries.country == 'AL') |
                                    (countries.country == 'AM') |
                                    (countries.country == 'BA') |
                                    (countries.country == 'FO') |
                                    (countries.country == 'GE') |
                                    (countries.country == 'XK') |
                                    (countries.country == 'MD') |
                                    (countries.country == 'ME') |
                                    (countries.country == 'MK') |
                                    (countries.country == 'NO') |
                                    (countries.country == 'RS') |
                                    (countries.country == 'TN') |
                                    (countries.country == 'TR') |
                                    (countries.country == 'UA'), 1, 0)


countries['ALL_Widening'] = np.where((countries.EU_Widening == 1) | (countries.not_EU_Widening == 1), 1, 0)
countries
```

```python
countries2 = countries.loc[:, ~countries.columns.isin(['CD_APPL_PIC', 'country'])]

countries2 = countries2.groupby(['project_id']).max().reset_index()

countries2 = countries2.rename(columns={'US': 'one_country_is_US',
                                        'UK': 'one_country_is_UK',
                                        'KR': 'one_country_is_SouthKorea',
                                        'JP': 'one_country_is_Japan',
                                        'CH': 'one_country_is_China',
                                        'EU27': 'at_least_one_EU27',
                                        'EU_Widening': 'at_least_one_EU_Widening',
                                        'not_EU_Widening': 'at_least_one_not_EU_Widening',
                                        'ALL_Widening': 'at_least_one_ALL_Widening'})

countries2
```

```python
#share of countries

n_countries = countries.groupby(['project_id'])['country'].nunique().reset_index()
n_countries = n_countries.rename(columns={'country': 'n_countries'})

#count total number of countries
total_number_of_countries = countries['country'].nunique()
n_countries['share_of_countries'] = round((n_countries['n_countries']/total_number_of_countries), 4)

print(total_number_of_countries)
n_countries
```

```python
###############################################################################################
```

```python
data = pd.merge(proj_character, n_countries, how='inner', on='project_id')

data = pd.merge(data, countries2, how='inner', on='project_id')

data = pd.merge(data, dummies_cd_org_type, how='inner', on='project_id')

data
```

```python
#save file project_id_technote_industry 
data_dir = '/Users/davidebonaglia/Dropbox/Projects UNU/Excellent Science/data_cleaned/'
data.to_excel(data_dir + 'participants_cleaned.xlsx', index=False)
```
