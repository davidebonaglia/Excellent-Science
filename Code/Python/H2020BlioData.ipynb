```python
#filter out CD_PROJECT_ID_01 and order the columns
H2020BiblioData2 = H2020BiblioData[[' Digital Object Identifier (DOI)', 'Publication Year', 
                                    'CD_PROJECT_ID_01', 'programme_PROJ_ID_01', 'programmes_PROJ_ID_01', '_mv_CD_ACTION_TYPE_PROJ_ID_01',
                                    'start_date_PROJ_ID_01', 'end_date_PROJ_ID_01',
                                    ' Funding Agency and Grant Number', 'Research Area', 'Interdisciplinary', 
                                    'Document Type', 'Times Cited', 'Field Expected Citations', 'Field Normalized Citation Impact', 
                                    'Percentile in Subject Area', 'Top 1%',  ' Open Access Indicator']]

#rename columns
H2020BiblioData2 = H2020BiblioData2.rename(columns={' Digital Object Identifier (DOI)': 'publication',
                                                    'Publication Year': 'publn_year',
                                                    'CD_PROJECT_ID_01' : 'project_id',
                                                    'programme_PROJ_ID_01': 'progr_proj_id',
                                                    '_mv_CD_ACTION_TYPE_PROJ_ID_01': 'action_type_proj_id',
                                                    'start_date_PROJ_ID_01' : 'starting_date',
                                                    'end_date_PROJ_ID_01' : 'ending_date',
                                                    ' Funding Agency and Grant Number': 'fund_agency',
                                                    'Research Area': 'research_areas',
                                                    'Document Type': 'document_type',
                                                    'Times Cited': 'times_cited',
                                                    'Field Expected Citations' : 'field_expected_citations',
                                                    'Field Normalized Citation Impact': 'field_normalized_cit_impact',
                                                    'Top 1%': 'top_1',
                                                    ' Open Access Indicator': 'open_access',
                                                    'Percentile in Subject Area':'perc_subj_area'})

#select only the articles and proceeding papers
H2020BiblioData2 = H2020BiblioData2.loc[(H2020BiblioData2.document_type == 'Article') |
                                        (H2020BiblioData2.document_type == 'Proceedings Paper')]

H2020BiblioData2 = H2020BiblioData2.sort_values(by=['publication', 'publn_year'], ascending=True)

H2020BiblioData2
```

```python
import numpy as np
import datetime

period = H2020BiblioData2[['project_id', 'starting_date', 'ending_date']]
period = period.sort_values(by=['project_id'], ascending=True)
period = period.drop_duplicates()


period['period_in_days'] = ((period.ending_date - period.starting_date)/np.timedelta64(1, 'D'))
period['period_in_days'] = period['period_in_days'].astype(int)

period['period_in_months'] = ((period.ending_date - period.starting_date)/np.timedelta64(1, 'M'))
period['period_in_months'] = period['period_in_months'].astype(int)

period['starting_date'] =  pd.to_datetime(period['starting_date'], format='%Y-%m-%d')
period['starting_date'] = period['starting_date'].dt.strftime('%Y-%m-%d')

period['ending_date'] =  pd.to_datetime(period['ending_date'], format='%Y-%m-%d')
period['ending_date'] = period['ending_date'].dt.strftime('%Y-%m-%d')

period
```

```python
fund_agency = H2020BiblioData2[['publication', 'project_id', 'fund_agency']]
fund_agency = fund_agency.loc[fund_agency.fund_agency != 0]

fund_agency = fund_agency.sort_values(by=['project_id'], ascending=True)
fund_agency
```

```python
#clean fund_agency

#trim leading and trailing blank spaces
fund_agency['fund_agency_cleaned'] = fund_agency['fund_agency'].str.lstrip() 
fund_agency['fund_agency_cleaned'] = fund_agency['fund_agency_cleaned'].str.rstrip()
#put all strings in uppercase
fund_agency['fund_agency_cleaned'] = fund_agency['fund_agency_cleaned'].str.upper() 
#remove non-alphanumeric chars
fund_agency['fund_agency_cleaned'] = fund_agency['fund_agency_cleaned'].str.replace('\W', ' ') 
#remove numbers
fund_agency['fund_agency_cleaned'] = fund_agency['fund_agency_cleaned'].str.replace('\d+', '')
#remove double spaces
fund_agency['fund_agency_cleaned'] = fund_agency['fund_agency_cleaned'].replace(r'\s+', ' ', regex=True)

fund_agency
```

```python
#find main categories of funding agency based on fund_agency_cleaned's string

#University
fund_agency['fund_agency_University'] = np.where(((fund_agency.fund_agency_cleaned.str.contains("UNIVERSITY")) |
                                                    (fund_agency.fund_agency_cleaned.str.contains("SCHOOL")) |
                                                    (fund_agency.fund_agency_cleaned.str.contains("DOCTORAL")) |
                                                    (fund_agency.fund_agency_cleaned.str.contains("ACADEMY")) |
                                                    (fund_agency.fund_agency_cleaned.str.contains("LEUVEN")) |
                                                    (fund_agency.fund_agency_cleaned.str.contains("COLLEGE"))) , 1, 0)

#Ministry
fund_agency['fund_agency_Ministry'] = np.where(((fund_agency.fund_agency_cleaned.str.contains("MINISTRY")) |
                                            (fund_agency.fund_agency_cleaned.str.contains("MINISTERE")) |
                                            (fund_agency.fund_agency_cleaned.str.contains("MINISTERIO")) |
                                            (fund_agency.fund_agency_cleaned.str.contains("MINISTRO"))) , 1, 0)

#Research Institute
fund_agency['fund_agency_Research_Institute'] = np.where(((fund_agency.fund_agency_cleaned.str.contains("RESEARCH COUNCIL")) |
                                                            (fund_agency.fund_agency_cleaned.str.contains("RESEARCH INSTITUTE")) |
                                                            (fund_agency.fund_agency_cleaned.str.contains("JOINT RESEARCH CENTRE")) |
                                                            (fund_agency.fund_agency_cleaned.str.contains("JOINT RESEARCH CENTER")) |
                                                            (fund_agency.fund_agency_cleaned.str.contains("RESEARCH FOUNDATION")) |
                                                            (fund_agency.fund_agency_cleaned.str.contains("RESEARCH DEVELOPMENT")) |
                                                            (fund_agency.fund_agency_cleaned.str.contains("RESEARCH AGENCY")) |
                                                            (fund_agency.fund_agency_cleaned.str.contains("RECHERCHE SCIENTIFIQUE")) |
                                                            (fund_agency.fund_agency_cleaned.str.contains("SCIENTIFIC RESEARCH")) |
                                                            (fund_agency.fund_agency_cleaned.str.contains("INSTITUTE"))) , 1, 0)

fund_agency['fund_agency_Others'] = np.where(((fund_agency.fund_agency_University == 0) &
                                                (fund_agency.fund_agency_Ministry == 0) &
                                                (fund_agency.fund_agency_Research_Institute == 0) &
                                                (fund_agency.fund_agency_cleaned.notna())), 1, 0)

fund_agency 
```

```python
#Number of Funding Agencies

#to count how many agencies per proj-publ, we approximate counting the number of ";" in the string
fund_agency['n_agencies'] = fund_agency.fund_agency.str.count(';')
#however, since at the end of each string, ";" is missing, we have to add 1 to count the exact number of agencies
fund_agency['n_agencies'] = fund_agency['n_agencies'] + 1
#for the same reason, we have to add an 1 to those proj-publ in which n_org=0
fund_agency['n_agencies'] = np.where((fund_agency.n_agencies == 0), 1, fund_agency.n_agencies)
fund_agency['n_agencies'] = fund_agency['n_agencies'].astype(int)

fund_agency
```

```python
at_least_one_fund_agency = fund_agency.groupby(['project_id']).agg({'fund_agency_University':'max',
                                                                    'fund_agency_Ministry':'max',
                                                                    'fund_agency_Research_Institute':'max',
                                                                    'fund_agency_Others':'max',
                                                                    'n_agencies':'sum'}).reset_index()

at_least_one_fund_agency = at_least_one_fund_agency.rename(columns={'fund_agency_University': 'at_least_one_fund_agency_University',
                                                                    'fund_agency_Ministry': 'at_least_one_fund_agency_Ministry',
                                                                    'fund_agency_Research_Institute': 'at_least_one_fund_agency_ResInst',
                                                                    'fund_agency_Others': 'at_least_one_fund_agency_Others'})

at_least_one_fund_agency = at_least_one_fund_agency.sort_values(by=['project_id'], ascending=True)
at_least_one_fund_agency
```
