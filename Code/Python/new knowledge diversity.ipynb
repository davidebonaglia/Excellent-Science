```python
import pandas as pd

data_dir = '/Users/davidebonaglia/Dropbox/Projects UNU/Excellent Science/'
H2020BiblioData = data_dir + 'H2020BiblioData&Indicators_ExcelScie_ResEuro_ProjIDs_v1.3.1.xlsx'

H2020BiblioData = pd.read_excel(H2020BiblioData)
H2020BiblioData
```

```python
#filter out CD_PROJECT_ID_01 and order the columns
H2020BiblioData2 = H2020BiblioData[[' Digital Object Identifier (DOI)', 'Publication Year', 
                                    'CD_PROJECT_ID_01', 'programme_PROJ_ID_01', 'programmes_PROJ_ID_01', '_mv_CD_ACTION_TYPE_PROJ_ID_01',
                                    'start_date_PROJ_ID_01', 'end_date_PROJ_ID_01',
                                    ' Funding Agency and Grant Number', 'Research Area', 'Interdisciplinary', 
                                    'Document Type', 'Times Cited', 'Field Expected Citations', 'Field Normalized Citation Impact', 
                                    'Percentile in Subject Area', 'Top 1%',  ' Open Access Indicator']]

#rename columns
H2020BiblioData2 = H2020BiblioData2.rename(columns={' Digital Object Identifier (DOI)': 'publication',
                                                    'Publication Year': 'publn_year',
                                                    'CD_PROJECT_ID_01' : 'project_id',
                                                    'programme_PROJ_ID_01': 'progr_proj_id',
                                                    '_mv_CD_ACTION_TYPE_PROJ_ID_01': 'action_type_proj_id',
                                                    'start_date_PROJ_ID_01' : 'starting_date',
                                                    'end_date_PROJ_ID_01' : 'ending_date',
                                                    ' Funding Agency and Grant Number': 'fund_agency',
                                                    'Research Area': 'research_areas',
                                                    'Document Type': 'document_type',
                                                    'Times Cited': 'times_cited',
                                                    'Field Expected Citations' : 'field_expected_citations',
                                                    'Field Normalized Citation Impact': 'field_normalized_cit_impact',
                                                    'Top 1%': 'top_1',
                                                    ' Open Access Indicator': 'open_access',
                                                    'Percentile in Subject Area':'perc_subj_area'})

#select only the articles and proceeding papers
H2020BiblioData2 = H2020BiblioData2.loc[(H2020BiblioData2.document_type == 'Article') |
                                        (H2020BiblioData2.document_type == 'Proceedings Paper')]

H2020BiblioData2 = H2020BiblioData2.sort_values(by=['publication', 'publn_year'], ascending=True)

H2020BiblioData2
```

```python
research_areas = H2020BiblioData2[['project_id', 'research_areas']]
research_areas = research_areas.sort_values(by=['project_id'], ascending=True)

research_areas
```

```python
df = research_areas.research_areas.str.split(';', expand=True).stack().reset_index(level=1, drop=True).to_frame('research_areas').join(research_areas[['project_id']], how='left')

df = df.sort_values(by=['project_id'], ascending=True)
df
```

```python
knowledge_diversity = df.groupby(['project_id'])['research_areas'].nunique().reset_index()

knowledge_diversity = knowledge_diversity.rename(columns={'research_areas': 'knowledge_diversity_3'})
knowledge_diversity
```

```python
#save file  
data_dir = '/Users/davidebonaglia/Dropbox/Projects UNU/Excellent Science/data_cleaned/'
knowledge_diversity.to_excel(data_dir + 'knowledge_diversity_29122022.xlsx', index=False)
```
